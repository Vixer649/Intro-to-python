```python
# TASK 6
dict = {}
for i in range(5):
    dict[i+1] = {}
    print("Enter Details of student "+str(i+1))
    n= input("Enter name ")
    roll = input("Enter roll number ")
    gpa =  input("Enter GPA ")

    dict[i+1]["name"] = n
    dict[i+1]["roll_number"] = roll
    dict[i+1]["GPA"] = gpa

for key,val in dict.items():
    print(key," : ",val)

num = int(input("Enter the number of student to see the details"))

for i in dict:
    if i == num:
        print(dict[i])
```
